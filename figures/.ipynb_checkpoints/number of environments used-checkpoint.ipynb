```python
def get_predictions(this_data,train_conditions,test_conditions,training_bcs,testing_bcs,full_model):
    

    all_conditions = list(train_conditions) + list(test_conditions)
#     all_conditions = train_conditions + test_conditions

    new_train_locs = np.where(np.isin(all_conditions,train_conditions))[0]
    new_test_locs = np.where(np.isin(all_conditions,test_conditions))[0]
    
    training_bcs = list(training_bcs)
    testing_bcs = list(testing_bcs)

    # all_bcs = sorted(list(dataset['training_bcs'])+list(dataset['testing_bcs']))
#     all_bcs = sorted(list(minimal_training_bcs + minimal_testing_bcs))
    all_bcs = sorted(list(training_bcs)+list(testing_bcs))
#     training_bcs = list(minimal_training_bcs)
#     testing_bcs = list(minimal_testing_bcs)

    these_bcs = testing_bcs
    these_genes = this_data[this_data['barcode'].isin(these_bcs)]['mutation_type'].values

#     ves = []

    train = [list(new_train_locs),list(np.where(np.isin(all_bcs,training_bcs))[0])]
    test = [list(new_test_locs),list(np.where(np.isin(all_bcs,testing_bcs))[0])]

    train_c = train[0]
    train_m = train[1]

    test_c = test[0]
    test_m = test[1]

    full_model = range(full_model)
    
    ves = np.zeros((len(test_conditions),len(full_model)))
    overall_ve = np.zeros(len(full_model))
    
    print(overall_ve.shape)

    this_fitness =  this_data[this_data['barcode'].isin(all_bcs)][all_conditions].values
    
    for this_m in range(1,len(full_model)+1):

        this_model = list(range(this_m))

        ve, old_mut_locs, new_mut_locs, old_cond_locs, new_cond_locs, these_sigmas = tools.SVD_mixnmatch_locations(this_fitness,train,test,this_model)

        these_sigmas = these_sigmas[:len(this_model),:len(this_model)]
        
        this_f = this_fitness[test_m,:][:,test_c]
#         print(this_f.shape)
        this_estimates = np.dot(new_mut_locs,np.dot(these_sigmas,new_cond_locs.T))
        
        this_ve = tools.var_explained_weighted_by_type(this_f,this_estimates,these_genes)

        overall_ve[this_m-1] = this_ve[0]
    
        for c,condition in enumerate(test_conditions):

            condition_loc = np.where(np.isin(all_conditions,condition))[0][0]

            if condition in test_conditions:
                this_estimates = [np.dot(new_mut_locs,np.dot(these_sigmas,new_cond_locs.T))[bc,condition_loc-old_cond_locs.shape[0]] for bc in range(len(these_bcs))]
            else:
                this_estimates = [np.dot(new_mut_locs,np.dot(these_sigmas,old_cond_locs.T))[bc,condition_loc] for bc in range(len(these_bcs))]    

            this_f = this_fitness[test_m,len(train_conditions)+c]

            this_ve = tools.var_explained_weighted_by_type(this_f,this_estimates,these_genes)

            ves[c,this_m-1] = this_ve[0]
            
    return ves,overall_ve
```

```python
np.random.seed(9282020)

n_cols_chosen = range(10,25+1)
# n_cols_chosen = [10,15,20,25]
n_sets = 25

test_conditions = list(sorted_nonm3_cols[first_nonsubtle:])


training_conditions = list(sorted_m3_cols) + list(sorted_nonm3_cols[:first_nonsubtle])

max_d_array = []
# prediction_array = np.zeros((len(n_cols_chosen),n_sets,len(test_conditions)))
prediction_array = []
overall_prediction = []
for n,n_cols in enumerate(n_cols_chosen):
    already_chosen = []
    max_d_array.append([])
    prediction_array.append([])
    overall_prediction.append([])
 
    
    if n_cols in [24,25]:
        for i,chosen_ones in enumerate(combinations(range(len(training_conditions)),n_cols)):
            these_cols = np.asarray(training_conditions)[list(chosen_ones)]
            
            this_fitness = this_data[this_data['barcode'].isin(list(minimal_training_bcs)+list(minimal_testing_bcs))][these_cols].values
            this_error = this_data[this_data['barcode'].isin(list(minimal_training_bcs)+list(minimal_testing_bcs))][[col.replace('_fitness','_error') for col in these_cols ]].values

            max_d, s, var_e_s= tools.svd_noise_detection(this_fitness,this_error,100)

            ves,overall_ve = get_predictions(this_data,these_cols,test_conditions,minimal_training_bcs,minimal_testing_bcs,max_d)

            max_d_array[n].append(max_d)
            prediction_array[n].append(ves)
            overall_prediction[n].append(overall_ve)
        
    else:
        for i in range(n_sets):
    
            these_cols = np.random.choice(training_conditions,n_cols,replace=False)
            x = 0
            while these_cols in already_chosen:
                print('oops I made it here',x)
                x += 1
                these_cols = np.random.choice(training_conditions,n_cols,replace=False)

            this_fitness = this_data[this_data['barcode'].isin(list(minimal_training_bcs)+list(minimal_testing_bcs))][these_cols].values
            this_error = this_data[this_data['barcode'].isin(list(minimal_training_bcs)+list(minimal_testing_bcs))][[col.replace('_fitness','_error') for col in these_cols ]].values

            max_d, s, var_e_s= tools.svd_noise_detection(this_fitness,this_error,100)
            
            ves,overall_ve = get_predictions(this_data,these_cols,test_conditions,minimal_training_bcs,minimal_testing_bcs,max_d)

        
            max_d_array[n].append(max_d)
            prediction_array[n].append(ves)
            overall_prediction[n].append(overall_ve)
        
```
